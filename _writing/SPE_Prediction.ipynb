```python
import numpy as np

# 랜덤시드 고정시키기
np.random.seed(5)

# 데이터셋 불러오기
dataset = np.loadtxt("./warehouse/data_binary.csv", delimiter=",")

# split into train and test sets 
train_size = int(len(dataset) * 7.0/10.0 * 4.0/5.0)
val_size = int(len(dataset) * 7.0/10.0 * 1.0/5.0)    
test_size = len(dataset) - train_size - val_size
train = dataset[0:train_size,:]
val = dataset[train_size:(train_size+val_size),:]
test = dataset[(train_size+val_size):len(dataset),:]

print('train:'+str(len(train)) + ' val:'+str(len(val)) + ' test:'+str(len(test)))
```

```python
# 입력(X)과 출력(Y) 변수로 분리하기
trainX = train[:,0:4]
trainY = train[:,4]

valX = val[:,0:4]
valY = val[:,4]

testX = test[:,0:4]
testY = test[:,4]
```

```python
from keras.models import Sequential
from keras.layers import Dense

# 모델 구성하기
model = Sequential()
model.add(Dense(12, activation='relu', kernel_initializer='uniform', input_dim=4))
model.add(Dense(8, activation='relu', kernel_initializer='uniform'))
model.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))

# 모델 엮기
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 학습시키기
      
model.fit(trainX, trainY, nb_epoch=100, batch_size=10, validation_data=(valX, valY), shuffle=True)
```

```python
from sklearn.metrics import classification_report

# 모델 평가하기
scores = model.evaluate(testX, testY)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))

predicted = model.predict(testX)

for i in range(len(testX)):
    if testY[i] == 1:
        print(predicted[i])
```
