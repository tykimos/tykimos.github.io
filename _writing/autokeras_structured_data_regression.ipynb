```python
# 1. 오토케라스 설치
!pip install autokeras

# 2. 패키지 가져오기
import numpy as np
import tensorflow as tf
import pickle
import pandas as pd
import autokeras as ak

# 3. 데이터 불러오기
x_train = pd.read_pickle('X_train.pkl').to_numpy()
y_train = pd.read_pickle('Y_train.pkl').to_numpy()
x_test = pd.read_pickle('X_test.pkl').to_numpy()
y_test = pd.read_pickle('Y_test.pkl').to_numpy()

# 4. 모델 정의
reg = ak.StructuredDataRegressor(max_trials=500, overwrite=True)

# 5. 모델 학습
reg.fit(x_train, y_train, epochs=200)

# 6. 모델 평가
result = reg.evaluate(x_test, y_test)
print(result)
```
